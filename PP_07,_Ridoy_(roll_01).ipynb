```python
#(2) Visualizing Bar Graph

#(We use bar graph for visualizing data: compare data across categories,highlight patterns and trends,analyze aggregated data,data interpretation )

import pandas as pd

import matplotlib.pyplot as plt

url="https://raw.githubusercontent.com/ridoy365/Python-Programming-07-Roll-01/refs/heads/main/environmental_data.csv"

df=pd.read_csv(url)

temperature_sum=df["Temperature"].sum()

humidity_sum=df["Humidity"].sum()

pressure_sum=df["Pressure"].sum()

y=[temperature_sum, humidity_sum, pressure_sum]
x=["Temperature","Humidity","Pressure"]

plt.figure(figsize=(8,4))

plt.bar(x,y, color="purple")

plt.grid(color="black",linestyle="-.",linewidth=".60")

plt.title("Distribution of Environmental Data", fontweight="bold", color="black")

plt.xlabel("Environmental Variables", fontsize=10, fontweight="bold", color="black")

plt.ylabel("Sum of Values", fontsize=10, fontweight="bold", color="black")

plt.show()
```

```python
#(3) Visualizing Pie Chart

#(We use pie chart for: show proportional relationships,easy interpretation,highlight dominance )

import pandas as pd

import matplotlib.pyplot as plt

url="https://raw.githubusercontent.com/ridoy365/Python-Programming-07-Roll-01/refs/heads/main/environmental_data.csv"

df=pd.read_csv(url)

temperature_sum=df["Temperature"].sum()

humidity_sum=df["Humidity"].sum()

pressure_sum=df["Pressure"].sum()

y=[temperature_sum, humidity_sum, pressure_sum]

x=["Temperature", "Humidity", "Pressure"]

plt.pie(y, labels=x, autopct="%1.1f%%", startangle=91, radius=1, explode=[.1,0,0])


plt.title("Distribution of Environmental Variables", fontsize="10", fontweight="bold", color="black")

plt.show()
```

```python
# (4) Visualizing Line Graph

#(we use line graph for: show trend over time,continuous data representation, patterns of data, prediction, compare multiple time series data)

import pandas as pd

import matplotlib.pyplot as plt

url="https://raw.githubusercontent.com/ridoy365/Python-Programming-07-Roll-01/refs/heads/main/environmental_data.csv"

df=pd.read_csv(url)

humidity= df["Humidity"]

pressure=df["Pressure"]

temperature=df["Temperature"]

plt.figure(figsize=(9,5))

plt.plot(
    humidity, temperature,
    label="Temperature & Humidity", color="teal",
    ls="solid", marker="o"
    )

plt.plot(
    pressure, temperature,
    label="Temperature & Pressure", color="green",
    ls="dotted", marker="o"
    )

plt.grid(color="black", linestyle="--", linewidth=0.6)

plt.title("Temperature, Humidity and Pressure Relation", fontsize=12, fontweight="bold", color="black")

plt.xlabel("Humidity/ Pressure", fontweight="bold", color="black")

plt.ylabel("Temperature", fontweight="bold", color="black")

plt.show()
```

```python
# (5) Visualizing Bar Graphs with Seaborn Library

#(for data visualization we use seaborn library in python since it provides higher level interface for creating attractive and informative statistical graph)

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

url="https://raw.githubusercontent.com/ridoy365/Python-Programming-07-Roll-01/refs/heads/main/environmental_data.csv"

df=pd.read_csv(url)

temperature_sum=df["Temperature"].sum()

humidity_sum=df["Humidity"].sum()

pressure_sum=df["Pressure"].sum()

y=[temperature_sum, humidity_sum, pressure_sum]
x=["Temperature", "Humidity", "Pressure"]

plt.figure(figsize=(8,4))

sum_df=pd.DataFrame({"factor":x, "sum":y})

sns.set(style="whitegrid")

sns.set(palette="pastel")

sns.set(font_scale=1.2)

sns.barplot(x="factor", y="sum", data=sum_df, color="brown")

plt.title("Distribution of Environmental Data", fontsize=12, fontweight="bold", color="black")

plt.xlabel("Environmental Variables", fontsize=10, fontweight="bold", color="black")

plt.ylabel("Sum of Values", fontsize=10, fontweight="bold", color="black")

plt.show()
```

```python
# (6) Visualizing Line Graph with Bokeh Library

#(we use bokeh library for creating interactive and web ready visualizations in python)

import pandas as pd

from bokeh.plotting import figure, show, output_notebook

from bokeh.models import HoverTool

url="https://raw.githubusercontent.com/ridoy365/Python-Programming-07-Roll-01/refs/heads/main/environmental_data.csv"

df=pd.read_csv(url)


temperature=df["Temperature"]

humidity=df["Humidity"]

pressure=df["Pressure"]


output_notebook()

p=figure (title="Temperature & Humidity", x_axis_label="Humidity", y_axis_label="Temperature")
q=figure (title="Temperature & Pressure", x_axis_label="Pressure", y_axis_label="Temperature")

p.line(
    humidity, temperature, line_width=3,
    color="yellow", legend_label="Temperature & Humidity"
    )

q.line(
    pressure, temperature, line_width=3,
    color="blue", legend_label="Temperature & Pressure"
    )

p.scatter(
    humidity, temperature, size=11, color="red",
    legend_label="Temperature & Humidity"
    )

q.scatter(
    pressure, temperature, size=11, color="black",
    legend_label="Temperature & Pressure"
    )

hover=HoverTool()

hover.tooltips=[("x","$x"),("y","$y")]

p.add_tools(hover)

show(p)

show(q)
```
